## Actual vs Predicted Sales

This graph compares actual sales with predicted sales generated using the Random Forest model. The close alignment indicates that the model captures sales patterns effectively.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../outputs/forecast.csv")
print(df.head())
```

```python
lt.figure(figsize=(10,5))

plt.plot(df["qty_sold"].tail(100).values, label="Actual")
plt.plot(df["predicted_sales"].tail(100).values, label="Predicted")

plt.legend()
plt.title("Actual vs Predicted Sales")

plt.savefig("../images/actual_vs_predicted.png")

plt.show()
```
